Set the settings here

```python
iter=100000 #number of simulations, do not increase as it might take a lot of time
coins=99 #coins to be donated (in the inventory after boss fight)
donated_already=200 # coins already donated to the machine by a character you are playing
res_arr=np.zeros(shape=int(iter))
for i in range(iter):
  donated=donated_already
  prob=0
  for j in range(coins):
    if np.random.binomial(1,prob)==1:
      res_arr[i]=donated-donated_already
      break
    donated=donated+1
    prob=np.floor(0.2*np.min([100,np.exp(0.023*donated)-1])+0.5)/100
  if res_arr[i]==0:
    res_arr[i]=coins



```

Graph of estimated distribution

```python
pex.histogram(res_arr,histnorm="percent")# sample of trials
```

Expected donation, standard deviation and 99.9% confidence interval for estimate. Note the expectation is far away from the mode.

```python
AVG=np.average(res_arr)
SE=(np.std(res_arr)/np.sqrt(iter))
ME=3.291*SE
conf_int=[AVG-ME,AVG+ME]
print("Average:"+str(AVG))
print("99% confidene interval:"+ str(conf_int))
print("Standard deviation:" +str(np.std(res_arr)))
```

```python
def prob_donated_X_or_more(X,res_arr):
  res_arr.sort()
  counter=0
  for element in res_arr:
    if element>=X:
      break
    counter=counter+1
  return 1-(counter/len(res_arr))
```

Change the number below to output the probability of donating the number or more coins

```python
print(prob_donated_X_or_more(50,res_arr))
```

```python
def estimate_distribution(support,res_arr):
  dist=np.zeros(shape=int(support))
  for i in range(support):
    dist[i]= prob_donated_X_or_more(i,res_arr)
  return dist
```

```python
dist=estimate_distribution(99,res_arr)
```

```python
def generate_sample(donated_already):
  iter=100
  coins=99
  res_arr=np.zeros(shape=int(iter))
  for i in range(iter):
    donated=donated_already
    prob=0
    for j in range(coins):
      if np.random.binomial(1,prob)==1:
        res_arr[i]=donated-donated_already
        break
      donated=donated+1
      prob=np.floor(0.2*np.min([100,np.exp(0.023*donated)-1])+0.5)/100
    if res_arr[i]==0:
      res_arr[i]=coins
  return res_arr


```

```python
avgs=np.zeros(shape=int(260/10))
for i in range(0,260,10):
  avgs[int(i/10)]=np.average(generate_sample(i))
  print(i)
```

Plot the expected value of donating in one run given the emount of coins you have already donated with a given character.

```python
pex.line(x=range(0,260,10),y=avgs)
```
